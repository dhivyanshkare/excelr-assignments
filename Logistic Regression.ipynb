## 3.Model Building

```python
features_train=train_data.drop('Survived',axis=1)
features_train
```

```python
target_train=train_data[['Survived']]
target_train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(features_train,target_train,train_size=0.8,random_state=11)
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred=logreg.predict(x_test)
y_pred
```

## 4.Model Evaluation

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score,classification_report,roc_curve,roc_auc_score,recall_score,precision_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
print('ACCURACY:',accuracy_score(y_test,y_pred)*100)
```

```python
roc_auc=roc_auc_score(y_test,y_pred)
print(roc_auc)
```

```python
fpr,tpr,thr=roc_curve(y_test,y_pred)
```

```python
plt.plot(fpr,tpr,label='logit model')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC Curve')
plt.plot([0,1],[0,1],color='red',linestyle='--')
plt.grid(False)
plt.show()
```

## 5.Interpretation:

```python
print("Intercept",logreg.intercept_)
print("slope",logreg.coef_)
```
