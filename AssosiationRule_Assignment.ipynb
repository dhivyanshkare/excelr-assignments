```python
transactions = cleaned_data.apply(lambda row: [item for item in row if item != ''], axis=1).tolist()
transactions[:5]
```

# **Association Rule Mining:**

```python
#!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te
```

```python
te_ary = te.fit(transactions).transform(transactions)
te_ary
```

```python
data = pd.DataFrame(te_ary, columns=te.columns_)
data.head()
```

```python
data.to_csv("data.csv", index=False)
```

```python
frequent_itemsets = apriori(data, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

# **Analysis and Interpretation:**

```python
def format_itemset(itemset):
    return ', '.join(sorted(list(itemset)))
```

```python
rules['antecedents_formatted'] = rules['antecedents'].apply(format_itemset)
rules['consequents_formatted'] = rules['consequents'].apply(format_itemset)
rules
```

```python
print("Top 10 Association Rules:")
rules[['antecedents_formatted', 'consequents_formatted', 'support', 'confidence', 'lift']]
```

# **Interview Questions:**

1.	**What is lift and why is it important in Association rules?**

* Lift, a key concept in association rule mining, measures the surprisingness or strength of an association rule.
* It indicates how much greater the probability of the right-hand side (consequent) is in the presence of the left-hand side (antecedent) compared to its probability in the absence of the antecedent.
* A higher lift value represents a more significant association, making it more important in decision-making processes.
* Lift helps identify the most relevant relationships between items, allowing businesses to optimize marketing strategies, product recommendations, and customer segmentation.

2.	**What is support and Confidence. How do you calculate them?**

* Support and confidence are fundamental concepts in decision-making and machine learning, particularly in classification and regression tasks.
* Support refers to the proportion of instances in the training dataset that belong to a certain class or category.
* Confidence, on the other hand, represents the certainty or reliability of a model's prediction.

**Calculation:**
* **Support :** counting the number of instances in each class and dividing by the total number of instances.

* **Confidence :** calculated by dividing the number of transactions containing both itemsets by the number of transactions containing the first itemset.
