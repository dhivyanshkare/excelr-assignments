### IQRs of all numerical columns

```python
#volume
Q1 = np.percentile(df['Volume'], 25)
Q3 = np.percentile(df['Volume'], 75)
IQR = Q3 - Q1

print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

```python
#avg price
Q1 = np.percentile(df['Avg Price'], 25)
Q3 = np.percentile(df['Avg Price'], 75) 
IQR = Q3 - Q1
print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

```python
#total sales
Q1 = np.percentile(df['Total Sales Value'], 25)
Q3 = np.percentile(df['Total Sales Value'], 75)
IQR = Q3 - Q1
print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

```python
#Discount Rate
Q1 = np.percentile(df['Discount Rate (%)'], 25)
Q3 = np.percentile(df['Discount Rate (%)'], 75)
IQR = Q3 - Q1
print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

```python
#Discount Amount
Q1 = np.percentile(df['Discount Amount'], 25)
Q3 = np.percentile(df['Discount Amount'], 75)
IQR = Q3 - Q1
print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

```python
# Net Sales Value
Q1 = np.percentile(df['Net Sales Value'], 25)   
Q3 = np.percentile(df['Net Sales Value'], 75)
IQR = Q3 - Q1
print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

### Standardization of Numerical Variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
plt.figure(figsize=(11, 7))
plt.title('Before Standardization')
for col in num_col.columns:
    sns.histplot(df[col],label=col)
plt.legend()
plt.show()
```

```python
scaler= StandardScaler()
df[num]=scaler.fit_transform(df[num])
df[num].head()
```

```python
plt.figure(figsize=(11, 7))
plt.title('After Standardization')
for col in num:
    sns.histplot(df[col],label=col)
plt.legend()
plt.show()
```

### Conversion of Categorical Data into Dummy Variables

```python
df_encod=pd.get_dummies(df, columns=cat, drop_first=True)
for col in df_encod.columns:
    if df_encod[col].dtype == 'bool':
        df_encod[col] = df_encod[col].astype('int')
df_encod.head()
```
